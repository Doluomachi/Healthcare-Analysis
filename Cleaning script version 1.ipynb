```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\PROBOOK 430 G2\Documents\Hospital_Data.csv")
```

```python
df.describe()
```

```python
#Filling the missing performance score with the mean of each department
df['Performance_Score'] = df['Performance_Score'].fillna(
    df.groupby('Department')['Performance_Score'].transform('mean'))
```

```python
df.info()
```

```python
#Finding the limits.
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
print('Upper_limit:', upper_limit)
print('Lower_limit:', lower_limit)
```

```python
#Finding the outliers
df.loc[(df['Salary'] > upper_limit)| (df['Salary'] < lower_limit)]
```

```python
df['Salary'] = df['Salary'].astype(float)
```

```python
#Capping the outliers.
df_capped = df
df_capped.loc[df_capped['Salary']> upper_limit, 'Salary'] = upper_limit
df_capped.loc[df_capped['Salary']< lower_limit, 'Salary'] = lower_limit
```

```python
#Dropping duplicates in Staff_id
df= df.drop_duplicates('Staff_ID')
```
